```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("/content/train.csv")
```

```python
class_count  = df['Pclass'].value_counts().sort_index()
class_count
```

```python
total_passenger = len(df)
print(total_passenger)
expected_counts = total_passenger/3

expected = [expected_counts]*3

print("Expected passenger count (assuming uniform distribution):\n",expected)
```

```python
# perform the chisquare goodness of fitness

chi2,p_value = chisquare(class_count,expected)

#print the results

print("\n Chisqaure Statistics:{:.2f}".format(chi2))
print("\n P-value Statistics:{:.4f}".format(p_value))
```

```python
# Interpret the results

alpha = 0.05

if p_value<alpha:
  print("\n We reject the null hypothesis. The distribution among the class is not uniform")
else:
  print("\n We accept the null hypothesis. The distribution among the class is  uniform")
```

```python
# we will use the chi-sqaure test for independence to see if the survival rate of passenger is independent of passenger class

import numpy as np
import pandas
from scipy.stats import contingency

data = pd.read_csv("/content/train.csv")
data.head()
```

```python
contingency_table = pd.crosstab(data['Survived'],data['Pclass'])
contingency_table
```

```python
#Perform the chi_square test
from scipy.stats import chi2_contingency
chi2,p_value,dof,expected = chi2_contingency(contingency_table)

#print the results

print("\nChisqaure Statistics:{:.2f}".format(chi2))
print("P-value :{:.4f}".format(p_value))
print("Degree of Freedom :{}".format(dof))
print("Expected Frequencies:\n{}".format(expected))
```

```python
#Interpreting the result

alpha = 0.05

if p_value<alpha:
  print("\n we reject the null hypothesis. There is a significant association between passenger class and survival rate")
else:
  print("\n we fail to reject the null hypothesis. There is a significant association between passenger class and survival rate")
```
